```python
%matplotlib inline
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 16, 12  # that's default image size for this interactive session
from IPython.display import display
import os
import sqlite3
import pandas
from pandas.tools.plotting import scatter_matrix
conn = sqlite3.connect('./../data/db.sqlite')
```

```python
COOKIE_ALL_BASIC = 'cookie_all_basic'  #クッキーユーザテーブル
DEV_TEST_BASIC = 'dev_test_basic'  #デバイスユーザテーブル
DEV_TRAIN_BASIC = 'dev_train_basic' #デバイスユーザテーブル
ID_ALL_IP = 'id_all_ip' #IPデバイステーブル
ID_ALL_PROPERTY = 'id_all_property' #サイトデバイステーブル
IPAGG_ALL = 'ipagg_all' #IPテーブル
PROPERTY_CATEGORY = 'property_category' #サイトテーブル

LIST_TABLES = [COOKIE_ALL_BASIC, DEV_TEST_BASIC, DEV_TRAIN_BASIC, IPAGG_ALL]
#ID_ALL_IP  ID_ALL_PROPERTY, PROPERTY_CATEGORY
```

```python
for table in LIST_TABLES:
    print table
    df = pandas.read_csv('../data/%s.csv'%table)
    df.to_sql(table, conn,
              if_exists='replace')
```

```python
def describe(data_name):
    df = pandas.read_csv('../data/%s.csv'%data_name)[:1000]
    desc = df.describe()
    display(desc)
    display(scatter_matrix(df[desc.columns]))
```
